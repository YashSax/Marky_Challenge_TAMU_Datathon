```python
test_data = pd.read_csv("D:/Vasudev_Agarwal/Work_Fun/1_Texas_AM/1_Fall_2023/Marky_Challenge_TAMU_Datathon/marky_data/test.csv")
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

features = ['user_edited', 'caption', 'parameters_chapter_summary', 'parameters_tone']

target = 'approved'

X = test_data[features]
#y = test_data[target]

test_data['caption_sentiment_label'] = [""] * len(test_data['caption'])
test_data['caption_sentiment_scores'] = [0] * len(test_data['caption'])

train_caption = test_data['parameters_chapter_summary'][7]
print(train_caption)
print(len(train_caption))


nltk.download("vader_lexicon")
analyzer = SentimentIntensityAnalyzer()

for i, caption in enumerate(test_data['caption']):
    sentiment = analyzer.polarity_scores(caption)
    compound_score = sentiment['compound']

    if compound_score >= 0.05:
        sentiment_label = 'Positive'
    elif compound_score <= -0.05:
        sentiment_label = "Negative"
    else:
        sentiment_label = "Neutral"

    test_data.loc[i, "caption_sentiment_label"] = sentiment_label
    test_data.loc[i, 'caption_sentiment_scores'] = compound_score
```

```python
test_data['chapter_summary_sentiment_label'] = [""] * len(test_data['caption'])
test_data['chapter_summary_sentiment_scores'] = [0] * len(test_data['caption'])




nltk.download("vader_lexicon")
analyzer = SentimentIntensityAnalyzer()

for i, caption in enumerate(test_data['parameters_chapter_summary']):
    if isinstance(caption, str):
        print("Heelo")
   
        sentiment = analyzer.polarity_scores(caption)
        compound_score = sentiment['compound']

        if compound_score >= 0.05:
            sentiment_label = 'Positive'
            print('positive')
        elif compound_score <= -0.05:
            sentiment_label = "Negative"
            print('negative')
        else:
            sentiment_label = "Neutral"
            print('neutral')
        print(sentiment_label)
        print(compound_score)
        test_data.loc[i, "chapter_summary_sentiment_label"] = sentiment_label
        test_data.loc[i, 'chapter_summary_sentiment_scores'] = compound_score
    else:
     
        test_data.loc[i, "chapter_summary_sentiment_label"] = "N/A"
        test_data.loc[i, 'chapter_summary_sentiment_scores'] = 0.0

```
